```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
netflix_df = pd.read_csv("C:/Users/riyas/Downloads/Netflix.csv.unknown")
print(netflix_df)

# Cleaning dataset
netflix_df.isnull().sum()
netflix_df.dropna()

# Creating 4 subplots: 2 rows, 2 columns
fig, axs = plt.subplots(2,2, figsize=(12,8))

axs[0,0].bar(netflix_df['Title'],netflix_df['Watch Hours (millions)'], color= 'seagreen')
axs[0,0].set_title("Netflix Shows vs Watch Hours")
axs[0,0].set_xlabel('Title')
axs[0,0].set_ylabel('Watch Hours (millions)')
axs[0,0].grid(color='gray',linestyle='--',linewidth=0.5)

genre_counts =netflix_df['Genre'].value_counts()
axs[0,1].pie(genre_counts,labels=genre_counts.index,startangle=90,colors=['skyblue','palegreen','lightcoral','gold','khaki'],autopct='%1.1f%%')
axs[0,1].set_title("Distribution of Shows by Genre")
axs[0,1].axis('equal')

axs[1,0].plot(netflix_df['Release Year'],netflix_df['Rating'],color='forestgreen',marker='o')
axs[1,0].set_title("Release Year vs Rating")
axs[1,0].set_xlabel("Release Year")
axs[1,0].set_ylabel("Rating")
axs[1,0].grid(True)

axs[1,1].scatter(netflix_df['Rating'],netflix_df['Watch Hours (millions)'],color='crimson')
axs[1,1].set_title("Rating vs Watch Hours (millions)")
axs[1,1].set_xlabel("Rating")
axs[1,1].set_ylabel("Watch Hours (millions)")
axs[1,1].grid(True)

plt.tight_layout()
plt.savefig("my_subplots.png")
plt.show()
        


    
```
